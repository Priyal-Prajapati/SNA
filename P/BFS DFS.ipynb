```python
# Degree in DFS
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()

graph.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])

graph.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('D', 'F'), ('E', 'D'), ('E', 'F'), ('F', 'D')])

pos = nx.spring_layout(graph)
nx.draw(graph, pos, with_labels=True)
plt.show()

print("Degree:", graph.degree())
print("In-degree:", graph.in_degree())
print("Out-degree:", graph.out_degree())

def dfs(graph, start):
    visited = set()
    traversal_order = []
    stack = [start]

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            traversal_order.append(node)
            for neighbor in graph.neighbors(node):
                if neighbor not in visited:
                    stack.append(neighbor)

    return traversal_order

# Perform DFS traversal starting from node 'A'
traversal_order = dfs(graph, 'A')
print("DFS Traversal Order:", traversal_order)
```
