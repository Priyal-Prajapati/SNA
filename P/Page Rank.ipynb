```python
import numpy as np
import pandas as pd
import networkx as nx

# Load dataset (assuming the file 'edges.csv' has two columns: 'source' and 'target')
df = pd.read_csv('edges.csv')

# Create directed graph from the dataset
graph = nx.DiGraph()
edges = list(zip(df['source'], df['target']))
graph.add_edges_from(edges)

# Number of nodes in the graph
n = graph.number_of_nodes()
nodes = list(graph.nodes)

# Initialize the adjacency matrix M
M = np.zeros((n, n))

# Fill the adjacency matrix M (stochastic matrix)
for i, node in enumerate(nodes):
    neighbors = list(graph.successors(node))
    if neighbors:
        for neighbor in neighbors:
            j = nodes.index(neighbor)
            M[j, i] = 1 / len(neighbors)  # Normalize by number of outgoing links

# Parameters
beta = 0.8  # Damping factor
limit = 1e-20  # Convergence threshold
A = beta * M + (1 - beta) * np.full((n, n), 1 / n)

# Initialize the rank vector (PageRank)
rj = np.full(n, 1 / n)

# Perform power iteration until convergence
while True:
    new_rj = np.dot(A, rj)
    if np.linalg.norm(new_rj - rj) < limit:
        break
    rj = new_rj

# Print the final PageRank vector
print("\nFinal PageRank:\n", dict(zip(nodes, rj)))
```

```python
import numpy as np
import pandas as pd
import networkx as nx

# Load dataset (assuming 'edges.csv' has two columns: 'source' and 'target')
df = pd.read_csv('edges.csv')

# Create a directed graph from the dataset
graph = nx.DiGraph()
edges = list(zip(df['source'], df['target']))
graph.add_edges_from(edges)

# Number of nodes
n = graph.number_of_nodes()
nodes = list(graph.nodes)

# Initialize the transition matrix (extra)
extra = np.zeros((n, n))

# Fill the transition matrix with probabilities
for i, node in enumerate(nodes):
    neighbors = list(graph.successors(node))
    if neighbors:
        for neighbor in neighbors:
            j = nodes.index(neighbor)
            extra[j, i] = 1 / len(neighbors)  # Normalize by the number of outgoing links

# Print the transition matrix
print("Transition matrix (extra):\n", extra)

# PageRank parameters
max_iterations = 100
limit = 1e-60  # Convergence threshold
rj = np.full(n, 1 / n)  # Initialize rank vector

# Perform power iteration until convergence
for _ in range(max_iterations):
    new_rj = np.dot(extra, rj)
    if np.linalg.norm(new_rj - rj) < limit:
        break
    rj = new_rj

# Print the final PageRank values
print("\nFinal PageRank:\n", dict(zip(nodes, rj)))
```
